```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("CC GENERAL.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.MINIMUM_PAYMENTS = df.MINIMUM_PAYMENTS.fillna(df.MINIMUM_PAYMENTS.mean())
df.CREDIT_LIMIT = df.CREDIT_LIMIT.fillna(df.CREDIT_LIMIT.mean())
```

```python
df = df.drop(['CUST_ID'], axis=1)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
#scaling dataframe
df_scaled = std_scaler.fit_transform(df)
```

```python
#Transforming the DataFrame into 2D DataFrame
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_ = pca.fit_transform(df_scaled)
pca_final = pd.DataFrame(data=pca_, columns=['component1', 'component2'])
pca_final
```

```python
#Finding the optimal number of clusters to use in the K_Means algorithm
from sklearn.cluster import KMeans
intertia = []
range_ = range(1,10)
for i in range_:
    km = KMeans(n_clusters=i)
    km.fit_predict(pd.DataFrame(df_scaled))
    intertia.append(km.inertia_)
plt.plot(range_, intertia, 'rx-')
plt.xlabel('range')
plt.ylabel('distortion')
plt.show()
```

```python
#number of clusters equals 4
#begin clustering
kmeans_model = KMeans(n_clusters=4, init='k-means++', random_state=0)
kmeans_model.fit_predict(df_scaled)
```

```python
pca_final_kmeans = pd.concat([pca_final, pd.DataFrame({'cluster':kmeans_model.labels_})], axis=1)
```

```python
plt.figure(figsize=(10,10))
ax = sns.scatterplot(x='component1', y='component2', hue='cluster', data=pca_final_kmeans, palette=['red', 'green', 'blue', 'yellow'])
plt.show()
```

```python
#adding cluster column to DataFrame
cluster_df = pd.concat([df, pd.DataFrame({'cluster':kmeans_model.labels_})], axis=1)
cluster_df
```

```python
#Visualising cluster's count
sns.countplot(x='cluster', data=cluster_df)
```

```python
#saving model
import joblib
joblib.dump(kmeans_model, 'kmeans_model.pkl')
```

```python
#saving cluster_df csv file
cluster_df.to_csv('clustered_customers.csv')
```

```python
#training and testing the kmeans_model
from sklearn.model_selection import train_test_split
X = cluster_df.drop(['cluster'], axis=1)
y = cluster_df.cluster
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_classifier = DecisionTreeClassifier(criterion='entropy')
model_classifier.fit(train_X, train_y)
```

```python
print(type(test_X))
```

```python
y_pred = model_classifier.predict(test_X)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(test_y, y_pred))
print(classification_report(test_y, y_pred))
```

```python
#saving the model
import pickle
filename = 'customer_classifier.sav'
pickle.dump(model_classifier, open(filename, 'wb'))
```

```python
#load the model
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(test_X, test_y)
print(result)
```
